```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from mlxtend.plotting import plot_decision_regions

### YOUR CODE HERE
lr = LogisticRegression(multi_class='multinomial', \
                                        solver='saga', \
                                        tol=1e-3, \
                                        max_iter=500) 
lr.fit(X_train, y_train)
y_lr = lr.predict(X_test)

svm = SVC(kernel='linear', C=1e5)
svm.fit(X_train, y_train)
y_svm = svm.predict(X_test)
```

```python
def plot_lr_decision_function(clf, X, Y):
    # Retrieve the model parameters.
    b = clf.intercept_[0]
    w1, w2 = clf.coef_.T
    # Calculate the intercept and gradient of the decision boundary.
    c = -b/w2
    m = -w1/w2

    # Plot the data and the classification with the decision boundary.
    xmin, xmax = min(X[:, 0]), max(X[:, 0])
    ymin, ymax = min(X[:, 1]), max(X[:, 1])
    xd = np.array([xmin, xmax])
    yd = m*xd + c
    plt.plot(xd, yd, 'k', lw=1, ls='--')
    plt.fill_between(xd, yd, ymin, color='tab:blue', alpha=0.2)
    plt.fill_between(xd, yd, ymax, color='tab:orange', alpha=0.2)

    plt.scatter(*X[Y==0].T, s=8, alpha=0.5)
    plt.scatter(*X[Y==1].T, s=8, alpha=0.5)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.ylabel(r'$x_2$')
    plt.xlabel(r'$x_1$')

    plt.show()
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=50, cmap='autumn')
plot_svc_decision_function(svm);
```

```python
skplt.metrics.plot_confusion_matrix(y_test, y_svm, normalize=True)
accuracy_score(y_test, y_svm)
```

```python
plot_lr_decision_function(lr, X_train, y_train)
```

```python
skplt.metrics.plot_confusion_matrix(y_test, y_lr, normalize=True)
accuracy_score(y_test, y_lr)
```

#### What have happenned with classification quality?

From Polynomial, Sigmoid and Radial Basis Function kernels, RBF is obviously the best because the accuracy score is 99%.

#### How did decision border changed for each kernel?

Each kernel completely changes the decision boundary based on the underlying algorithms used to determine the boundary. Polynomial simply tests a polynomial of degree X and then optimizes the decision boundary. One can play with plots of a n-dimensional decision boundary. RBG performed really well.

#### What degree have you chosen and why?

The degree setting only affects the polynomial kernels, it does not affect the others. Increasing the degree from 1 - 10 the accuracy score varied with each degree.

The final choice lands on the degree = 5, as this provides almost 90% accuracy score for polynomial features + SVM.

```python
def phi(X):
    """"Non Linear Transformation"""
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X1**2 + X2**2
    
    X_ = np.zeros((X.shape[0],3))
    print(X_.shape)
    
    X_[:,:-1] = X
    X_[:,-1] = X3

    return X_
```

```python
import plotly.express as px

def plot3d(X,show=True):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111,projection='3d')
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X[:,2]
    
    ax.scatter(X1,X2,X3,zdir='z',s=30,c=y_trainC,depthshade=True)
    plt.show()
    return ax
```

```python
X_highD = phi(X_trainC)
plot3d(X_highD)
```

```python
from sklearn.model_selection import cross_val_score

lr=LogisticRegression()

acc=cross_val_score(lr, X_highD, y_trainC,cv=5).mean()
print("Accuracy X(2D) is %.4f"%(acc*100))
```
